# Artifacts overview

```python
import pickle
import numpy as np
import pandas as pd
from pathlib import Path

PROCESSED_DIR = Path("../data/processed/full")

# No limit for columns
pd.set_option("display.max_columns", None)
```

```python
# Encoders
user_enc = pickle.load(open(PROCESSED_DIR / "user_enc.pkl", "rb"))
movie_enc = pickle.load(open(PROCESSED_DIR / "movie_enc.pkl", "rb"))

print("user_enc  to_idx:", list(user_enc["to_idx"].items())[:5])
print("user_enc  to_id: ", list(user_enc["to_id"].items())[:5])
print("movie_enc to_idx:", list(movie_enc["to_idx"].items())[:5])
```

```python
# Ratings splits
train_ratings = pickle.load(open(PROCESSED_DIR / "train_ratings.pkl", "rb"))
val_ratings = pickle.load(open(PROCESSED_DIR / "val_ratings.pkl", "rb"))
test_ratings = pickle.load(open(PROCESSED_DIR / "test_ratings.pkl", "rb"))
train_val_ratings = pickle.load(open(PROCESSED_DIR / "train_val_ratings.pkl", "rb"))

print(f"Train:     {len(train_ratings):,}")
print(f"Val:       {len(val_ratings):,}")
print(f"Test:      {len(test_ratings):,}")
print(f"Train+Val: {len(train_val_ratings):,}")
train_ratings.head()
```

```python
# Enriched movies
movies_enriched = pd.read_parquet(PROCESSED_DIR / "movies_enriched.parquet")
print(movies_enriched.shape)
print(movies_enriched.dtypes)
movies_enriched.head()
```

```python
# Genome tags mapping
genome_tags = pd.read_parquet(PROCESSED_DIR / "genome_tags.parquet")
genome_tags.head(10)
```

```python
# Genome matrix
genome_matrix = np.load(PROCESSED_DIR / "genome_matrix.npy")
print("Shape: ", genome_matrix.shape)
print("Dtype: ", genome_matrix.dtype)
print("Sample row (movie_idx=0):", genome_matrix[0, :10])
```
